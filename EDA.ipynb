## HyperSpectral Data

```python
wave_columns = list(filter(lambda col: 'Wave_' in col, df.columns))
len(wave_columns)
```

There appears to be 1018 wavelength columns of hyperspectral data. Rather than trying to look at individual columns, I am going to take the mean over a set of wavelengths. While this may loose some data on individual columns, this can provide us some interesting data about sets of wavelengths

```python
def bin_wavelengths(dataset, binspacing=100):
    output = {}
    wave_columns = list(filter(lambda col: 'Wave_' in col, dataset.columns))
    start = 500
    end = 1600
    end_current_range = start + binspacing
    while end_current_range <= end:
        bin_cols = [f'Wave_{wl}' for wl in list(range(start, end_current_range))]
        remove_cols = set(bin_cols) - set(dataset.columns)
        bin_cols = list(set(bin_cols) - set(remove_cols))
        means = df[bin_cols].mean(axis=1)
        output[f'{start}-{end_current_range}'] = means
        start = end_current_range
        end_current_range = start + binspacing
    return pd.DataFrame(output)
```

```python
binned_df = bin_wavelengths(df)
```

### Binned Distributions

```python
fig = plt.figure(figsize=(20,20))
for num, col in enumerate(binned_df):
    num+=1
    ax = fig.add_subplot(4,3,num)
    sns.histplot(data=binned_df, x=col, ax=ax)
title = plt.suptitle('Wavelength Value Distributions')
fig.tight_layout()
```

### Binned Correlations

```python
binned_df = binned_df.join(df['Measured_Chl'])
corr = binned_df.corr()
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
heat = sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Interestingly, it seems that **the smaller wavelengths (500-800) are strongly negatively correlated with our target**. The mid range wavelengths show small to no positive correlation, and the highest wavelengths show slight negative or no correlation

## Categorical Variables

## Write a function for comparing categorical variables

The following function can be applied to compare the distribution of the target relative to two categorical variables. This could help us understand any interesting relationships between categorical fields and the target. I also added a threshold for the hue category so as to keep the plots somewhat legible.There appears to be 1018 wavelength columns of hyperspectral data

```python
def describe_categorical(dataset, hue_column, facet, thresh=0.01):
    val_counts = dataset[hue_column].value_counts(normalize=True)
    hue_col_binned = f'{hue_column}_binned'
    dataset[hue_col_binned] = dataset.apply(lambda row: 'Other' if val_counts[row[hue_column]] < thresh else row[hue_column], axis=1 )
    
    sns.displot(data=dataset, x='Measured_Chl',  hue=hue_col_binned, col=facet, height=10)
```

```python
describe_categorical(df, 'Treatment', 'Site')
```

```python
describe_categorical(df, 'Genotype', 'Site')
```

```python
describe_categorical(df, 'Growing Season', 'Site')
```

```python
describe_categorical(df, 'Genotype', 'Treatment')
```

```python
describe_categorical(df, 'Genotype', 'Growing Season')
```
